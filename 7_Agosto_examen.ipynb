```python
import pandas as pd
df=pd.read_csv('/content/examen.csv')
```

```python
#En datos nulos observo que no hay, al realizar el codigo me genera error porque hay datos vacios, quise dejar mi codigo en error para que se mostrara
df['TotalCharges']=df['TotalCharges'].astype(int)
#codigo no valido porque hay filas con datos vacios o sin datos
```

```python
#A lo anterior dicho estoy remplazando los datos vacios
#Nuevo codigo para remplazar los datos vacios por nan
import numpy as np
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
```

```python
#con este codigo convierto mis datos a float
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
#comprobante de tipo de fila
df.info()
```

```python
#los datos nulos los podemos remplazar por 0 para no cargarle saldos adicionales a clientes que pueden que hayan cancelado su servicio
df = df.fillna(0)
```

```python
#comprobante del remplazo de nulos
df.isna().sum()
```

```python
#no existe columna con valores nulos
```

```python
df.describe().T
```

```python
df.dtypes
```

```python
#identificando que se remplacen las mayusculas y minusculas, en mi caso lo que revise no era necesario pero de igual forma quise hacer el ejercicio
df['gender'] = df['gender'].str.capitalize()
df['Partner'] = df['Partner'].str.capitalize()
df['Dependents'] = df['Dependents'].str.capitalize()
df['PhoneService'] = df['PhoneService'].str.capitalize()
df['MultipleLines'] = df['MultipleLines'].str.replace('No phone service', 'No Phone Service', case=False)
df['InternetService'] = df['InternetService'].str.capitalize()
df['OnlineSecurity'] = df['OnlineSecurity'].str.replace('No internet service', 'No Internet Service', case=False)
df['OnlineBackup'] = df['OnlineBackup'].str.replace('No internet service', 'No Internet Service', case=False)
df['DeviceProtection'] = df['DeviceProtection'].str.replace('No internet service', 'No Internet Service', case=False)
df['TechSupport'] = df['TechSupport'].str.replace('No internet service', 'No Internet Service', case=False)
df['StreamingTV'] = df['StreamingTV'].str.replace('No internet service', 'No Internet Service', case=False)
df['StreamingMovies'] = df['StreamingMovies'].str.replace('No internet service', 'No Internet Service', case=False)
df['Contract'] = df['Contract'].str.title()
df['PaperlessBilling'] = df['PaperlessBilling'].str.capitalize()
df['PaymentMethod'] = df['PaymentMethod'].str.title()
df['Churn'] = df['Churn'].str.capitalize()
```
